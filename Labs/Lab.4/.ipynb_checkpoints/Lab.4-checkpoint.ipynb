1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.count = 0
        self.max_value = max_value
        
    def increment(self):
        if self.count < self.max_value:
            self.count += 1
        else:
            print("Error: Counter has reached its maximum value.")
            
    def reset(self):
        self.count = 0
```

```python
my_counter = Counter(2)

print("Incrementing first time:")
my_counter.increment()
print("Current count:", my_counter.count)

print("\nIncrementing second time:")
my_counter.increment()
print("Current count:", my_counter.count)

print("\nIncrementing third time (should show error):")
my_counter.increment()

print("\nResetting counter:")
my_counter.reset()
print("Current count:", my_counter.count)
```

2. Copy and paste your solution to question 1 and modify it so that all the data held by the counter is private. Implement functions to check the value of the counter, check the maximum value, and check if the counter is at the maximum.

```python
class PrivateCounter:
    def __init__(self, max_value):
#Using _x_ makes data members private
        self.__count = 0
        self.__max_value = max_value
        
    def increment(self):
        if self.__count < self.__max_value:
            self.__count += 1
        else:
            print("Error: Counter has reached its maximum value.")
            
    def reset(self):
        self.__count = 0
        
#Checks the value of the counter
    def get_value(self):
        return self.__count
    
#Checks the maximum value
    def get_max_value(self):
        return self.__max_value
#Checks if the counter is at the maximum
    def is_at_max(self):
        return self.__count == self.__max_value
```

```python
my_private_counter = PrivateCounter(2)

print("Initial value:", my_private_counter.get_value())
print("Max value:", my_private_counter.get_max_value())
print("Is at max?", my_private_counter.is_at_max())

print("\nIncrementing twice...")
my_private_counter.increment()
my_private_counter.increment()

print("Current value:", my_private_counter.get_value())
print("Is at max?", my_private_counter.is_at_max())

print("\nAttempting to increment again:")
my_private_counter.increment()
```

```python
class Rectangle:
    def __init__(self, length, width, x, y):
# Making all data members private
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
        
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
# Accessors to retrieve the private values
    def get_length(self):
        return self.__length
        
    def get_width(self):
        return self.__width
        
    def get_x(self):
        return self.__x
        
    def get_y(self):
        return self.__y
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
import math

# 1. Base Class
class Shape:
    def area(self):
        raise NotImplementedError
    def perimeter(self):
        raise NotImplementedError
    def get_perimeter_points(self):
        raise NotImplementedError

# 2. Rectangle Class
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
        
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def get_length(self): return self.__length
    def get_width(self): return self.__width
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def get_perimeter_points(self):
        # Returns the 4 corners of the rectangle
        return [
            (self.get_x(), self.get_y()), 
            (self.get_x() + self.get_length(), self.get_y()), 
            (self.get_x() + self.get_length(), self.get_y() + self.get_width()), 
            (self.get_x(), self.get_y() + self.get_width()) 
        ]

# 3. Circle Class
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y
        
    def area(self):
        pi = 3.14159
        return pi * (self.__radius ** 2)
    
    def perimeter(self):
        pi = 3.14159
        return 2 * pi * self.__radius
    
    def get_radius(self): return self.__radius
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def get_perimeter_points(self):
        # Returns 16 points evenly spaced along the perimeter
        points = []
        pi = 3.14159
        
        for i in range(16):
            angle = (2 * pi / 16) * i
            point_x = self.get_x() + self.get_radius() * math.cos(angle)
            point_y = self.get_y() + self.get_radius() * math.sin(angle)
            points.append((point_x, point_y))
            
        return points

# 4. Triangle Class
class Triangle(Shape):
    def __init__(self, base, height, side1, side2, side3, x, y):
        self.__base = base
        self.__height = height
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        self.__x = x
        self.__y = y
        
    def area(self):
        return 0.5 * self.__base * self.__height
    
    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3
    
    def get_base(self): return self.__base
    def get_height(self): return self.__height
    def get_side1(self): return self.__side1
    def get_side2(self): return self.__side2
    def get_side3(self): return self.__side3
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def get_perimeter_points(self):
        # Returns 3 corners of the triangle
        return [
            (self.get_x(), self.get_y()), 
            (self.get_x() + self.get_base(), self.get_y()), 
            (self.get_x() + (self.get_base() / 2), self.get_y() + self.get_height()) 
        ]
```

```python
my_rect = Rectangle(10, 5, 0, 0)
print("Rectangle 4 points:\n", my_rect.get_perimeter_points())

my_tri = Triangle(base=10, height=5, side1=10, side2=7, side3=7, x=0, y=0)
print("\nTriangle 3 points:\n", my_tri.get_perimeter_points())

my_circ = Circle(5, 0, 0)
print("\nCircle 16 points:")
for pt in my_circ.get_perimeter_points():
    print(f"  ({pt[0]:.2f}, {pt[1]:.2f})")
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
import math

# 1. Base Class
class Shape:
    def area(self):
        raise NotImplementedError
    def perimeter(self):
        raise NotImplementedError
    def get_perimeter_points(self):
        raise NotImplementedError
    def is_inside(self, px, py):
        raise NotImplementedError

# 2. Rectangle Class
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
        
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def get_length(self): return self.__length
    def get_width(self): return self.__width
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def get_perimeter_points(self):
        return [
            (self.get_x(), self.get_y()), 
            (self.get_x() + self.get_length(), self.get_y()), 
            (self.get_x() + self.get_length(), self.get_y() + self.get_width()), 
            (self.get_x(), self.get_y() + self.get_width()) 
        ]
        
    def is_inside(self, px, py):
        in_x_bounds = self.get_x() <= px <= (self.get_x() + self.get_length())
        in_y_bounds = self.get_y() <= py <= (self.get_y() + self.get_width())
        return in_x_bounds and in_y_bounds

# 3. Circle Class
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y
        
    def area(self):
        pi = 3.14159
        return pi * (self.__radius ** 2)
    
    def perimeter(self):
        pi = 3.14159
        return 2 * pi * self.__radius
    
    def get_radius(self): return self.__radius
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def get_perimeter_points(self):
        points = []
        pi = 3.14159
        
        for i in range(16):
            angle = (2 * pi / 16) * i
            point_x = self.get_x() + self.get_radius() * math.cos(angle)
            point_y = self.get_y() + self.get_radius() * math.sin(angle)
            points.append((point_x, point_y))
            
        return points

    def is_inside(self, px, py):
        # (x2 - x1)^2 + (y2 - y1)^2 <= r^2
        dx = px - self.get_x()
        dy = py - self.get_y()
        distance_squared = (dx ** 2) + (dy ** 2)
        radius_squared = self.get_radius() ** 2
        return distance_squared <= radius_squared

# 4. Triangle Class
class Triangle(Shape):
    def __init__(self, base, height, side1, side2, side3, x, y):
        self.__base = base
        self.__height = height
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        self.__x = x
        self.__y = y
        
    def area(self):
        return 0.5 * self.__base * self.__height
    
    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3
    
    def get_base(self): return self.__base
    def get_height(self): return self.__height
    def get_side1(self): return self.__side1
    def get_side2(self): return self.__side2
    def get_side3(self): return self.__side3
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def get_perimeter_points(self):
        return [
            (self.get_x(), self.get_y()), 
            (self.get_x() + self.get_base(), self.get_y()), 
            (self.get_x() + (self.get_base() / 2), self.get_y() + self.get_height()) 
        ]

    def is_inside(self, px, py):
        x1, y1 = self.get_x(), self.get_y()
        x2, y2 = self.get_x() + self.get_base(), self.get_y()
        x3, y3 = self.get_x() + (self.get_base() / 2), self.get_y() + self.get_height()
        
        def calc_area(ax, ay, bx, by, cx, cy):
            return abs((ax * (by - cy) + bx * (cy - ay) + cx * (ay - by)) / 2.0)
        
        area1 = calc_area(px, py, x2, y2, x3, y3)
        area2 = calc_area(x1, y1, px, py, x3, y3)
        area3 = calc_area(x1, y1, x2, y2, px, py)
        
        total_sub_area = area1 + area2 + area3
        main_area = self.area()
        
# Checks if sub-areas equal main area (with small float tolerance)
        return abs(main_area - total_sub_area) < 0.0001
```

```python
my_rect = Rectangle(10, 5, 0, 0)
print("Point (5, 2) in Rectangle?", my_rect.is_inside(5, 2))   # True
print("Point (15, 2) in Rectangle?", my_rect.is_inside(15, 2)) # False

my_circ = Circle(5, 0, 0)
print("\nPoint (2, 2) in Circle?", my_circ.is_inside(2, 2))    # True
print("Point (6, 6) in Circle?", my_circ.is_inside(6, 6))      # False

my_tri = Triangle(base=10, height=5, side1=10, side2=7, side3=7, x=0, y=0)
print("\nPoint (5, 2) in Triangle?", my_tri.is_inside(5, 2))   # True
print("Point (0, 5) in Triangle?", my_tri.is_inside(0, 5))     # False
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
# 1. Updated Base Class
class Shape:
    def area(self):
        raise NotImplementedError
        
    def perimeter(self):
        raise NotImplementedError
        
    def get_perimeter_points(self):
        raise NotImplementedError
        
    def is_inside(self, px, py):
        raise NotImplementedError
        
    def overlaps(self, other_shape):
# 1. Check if any perimeter point of THIS shape is inside the OTHER shape
        for px, py in self.get_perimeter_points():
            if other_shape.is_inside(px, py):
                return True
                
# 2. Check if any perimeter point of the OTHER shape is inside THIS shape
        for px, py in other_shape.get_perimeter_points():
            if self.is_inside(px, py):
                return True
# If neither is true, they do not overlap
        return False

# 2. Rectangle Class
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
        
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def get_length(self): return self.__length
    def get_width(self): return self.__width
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def get_perimeter_points(self):
        return [
            (self.get_x(), self.get_y()), 
            (self.get_x() + self.get_length(), self.get_y()), 
            (self.get_x() + self.get_length(), self.get_y() + self.get_width()), 
            (self.get_x(), self.get_y() + self.get_width()) 
        ]
        
    def is_inside(self, px, py):
        in_x_bounds = self.get_x() <= px <= (self.get_x() + self.get_length())
        in_y_bounds = self.get_y() <= py <= (self.get_y() + self.get_width())
        return in_x_bounds and in_y_bounds

# 3. Circle Class
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y
        
    def area(self):
        pi = 3.14159
        return pi * (self.__radius ** 2)
    
    def perimeter(self):
        pi = 3.14159
        return 2 * pi * self.__radius
    
    def get_radius(self): return self.__radius
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def get_perimeter_points(self):
        points = []
        pi = 3.14159
        
        for i in range(16):
            angle = (2 * pi / 16) * i
            point_x = self.get_x() + self.get_radius() * math.cos(angle)
            point_y = self.get_y() + self.get_radius() * math.sin(angle)
            points.append((point_x, point_y))
            
        return points

    def is_inside(self, px, py):
        # (x2 - x1)^2 + (y2 - y1)^2 <= r^2
        dx = px - self.get_x()
        dy = py - self.get_y()
        distance_squared = (dx ** 2) + (dy ** 2)
        radius_squared = self.get_radius() ** 2
        return distance_squared <= radius_squared

# 4. Triangle Class
class Triangle(Shape):
    def __init__(self, base, height, side1, side2, side3, x, y):
        self.__base = base
        self.__height = height
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        self.__x = x
        self.__y = y
        
    def area(self):
        return 0.5 * self.__base * self.__height
    
    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3
    
    def get_base(self): return self.__base
    def get_height(self): return self.__height
    def get_side1(self): return self.__side1
    def get_side2(self): return self.__side2
    def get_side3(self): return self.__side3
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def get_perimeter_points(self):
        return [
            (self.get_x(), self.get_y()), 
            (self.get_x() + self.get_base(), self.get_y()), 
            (self.get_x() + (self.get_base() / 2), self.get_y() + self.get_height()) 
        ]

    def is_inside(self, px, py):
        x1, y1 = self.get_x(), self.get_y()
        x2, y2 = self.get_x() + self.get_base(), self.get_y()
        x3, y3 = self.get_x() + (self.get_base() / 2), self.get_y() + self.get_height()
        
        def calc_area(ax, ay, bx, by, cx, cy):
            return abs((ax * (by - cy) + bx * (cy - ay) + cx * (ay - by)) / 2.0)
        
        area1 = calc_area(px, py, x2, y2, x3, y3)
        area2 = calc_area(x1, y1, px, py, x3, y3)
        area3 = calc_area(x1, y1, x2, y2, px, py)
        
        total_sub_area = area1 + area2 + area3
        main_area = self.area()
        
# Checks if sub-areas equal main area (with small float tolerance)
        return abs(main_area - total_sub_area) < 0.0001
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
import math

# 1. Canvas Class (Copied exactly from Lecture 9)
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        # Ensure we don't draw outside the canvas and cause wrapping errors
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x, x+w):
            self.set_pixel(i, y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y, y+h):
            self.set_pixel(x, i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1, y2):
            x = int(slope * y)
            self.set_pixel(x, y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))


# 2. Shape Classes
class Shape:
    def __init__(self, name="", **kwargs):
        self.name = name
        self.kwargs = kwargs
    
    def paint(self, canvas): pass


class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
        
    def get_length(self): return self.__length
    def get_width(self): return self.__width
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def paint(self, canvas, char='*'):
        # Canvas treats 'x' as row and 'y' as col in its h_line/v_line logic
        canvas.v_line(self.get_y(), self.get_x(), self.get_width(), char=char)
        canvas.v_line(self.get_y(), self.get_x() + self.get_length() - 1, self.get_width(), char=char)
        canvas.h_line(self.get_y(), self.get_x(), self.get_length(), char=char)
        canvas.h_line(self.get_y() + self.get_width() - 1, self.get_x(), self.get_length(), char=char)


class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y
        
    def get_radius(self): return self.__radius
    def get_x(self): return self.__x
    def get_y(self): return self.__y
    
    def get_perimeter_points(self):
        points = []
        pi = 3.14159
        for i in range(32):
            angle = (2 * pi / 32) * i
            point_x = self.get_x() + self.get_radius() * math.cos(angle)
            point_y = self.get_y() + self.get_radius() * math.sin(angle)
            points.append((point_x, point_y))
        return points

    def paint(self, canvas, char='*'):
        for px, py in self.get_perimeter_points():
            canvas.set_pixel(int(round(py)), int(round(px)), char=char)

class Triangle(Shape):
    def __init__(self, base, height, side1, side2, side3, x, y):
        self.__base = base
        self.__height = height
        self.__x = x
        self.__y = y
        
    def get_base(self): return self.__base
    def get_height(self): return self.__height
    def get_x(self): return self.__x
    def get_y(self): return self.__y

    def paint(self, canvas, char='*'):
# 1. Calculate the exact mathematical peak (center) of the triangle
        base_y = self.get_y()
        top_y = self.get_y() - self.get_height()
        peak_x = self.get_x() + (self.get_base() - 1) / 2.0
        
        # Prevent division by zero if someone creates a flat triangle
        if self.get_height() == 0:
            for i in range(self.get_base()):
                canvas.set_pixel(base_y, self.get_x() + i, char=char)
            return

# 2. Draw perfectly interpolated left and right edges from top to bottom
        for i in range(self.get_height() + 1):
            current_y = top_y + i
            
            # What fraction of the way down are we? (0.0 at peak, 1.0 at base)
            fraction = i / float(self.get_height())
            
            # Calculate exactly where the left and right edges should be at this fraction
            left_x = peak_x - fraction * ((self.get_base() - 1) / 2.0)
            right_x = peak_x + fraction * ((self.get_base() - 1) / 2.0)
            
            # Plot the points (adding 0.5 safely rounds to the nearest integer pixel)
            canvas.set_pixel(current_y, int(left_x + 0.5), char=char)
            canvas.set_pixel(current_y, int(right_x + 0.5), char=char)
            
# 3. Draw the flat bottom base exactly to match the width
        for i in range(self.get_base()):
            canvas.set_pixel(base_y, self.get_x() + i, char=char)

    def paint(self, canvas, char='*'):
        # 1. Calculate the Y boundaries and the exact center X (peak) of the triangle
        base_y = self.get_y()
        top_y = self.get_y() - self.get_height()
        cx = self.get_x() + int(self.get_base() / 2)
        half_width = self.get_base() / 2.0
        
        # Prevent division by zero if someone creates a flat triangle
        if self.get_height() == 0:
            for i in range(self.get_base() + 1):
                canvas.set_pixel(base_y, self.get_x() + i, char=char)
            return

        # 2. Draw perfectly symmetric left and right edges from top to bottom
        for i in range(self.get_height() + 1):
            current_y = top_y + i
            # Calculate how far left/right to go from the center at this height
            x_offset = int((half_width * i) / self.get_height() + 0.5)
            
            # Plot the left and right points equally to guarantee perfect symmetry
            canvas.set_pixel(current_y, cx - x_offset, char=char)
            canvas.set_pixel(current_y, cx + x_offset, char=char)
            
        # 3. Draw the flat bottom base
        left_corner = cx - int(half_width)
        for i in range(self.get_base() + 1):
            canvas.set_pixel(base_y, left_corner + i, char=char)

    def paint(self, canvas, char='*'):
        # 1. Set the 3 corners. Notice we SUBTRACT height so the peak goes UP.
        x1, y1 = self.get_x(), self.get_y()
        x2, y2 = self.get_x() + self.get_base() - 1, self.get_y()
        x3, y3 = self.get_x() + int(self.get_base() / 2), self.get_y() - self.get_height()
        
        # 2. We use our own safe line-drawing algorithm to bypass the buggy Canvas.line
        def draw_safe_line(ax, ay, bx, by):
            dx, dy = bx - ax, by - ay
            steps = max(abs(dx), abs(dy))
            if steps == 0: return
            for i in range(steps + 1):
                px = ax + (dx * i) / steps
                py = ay + (dy * i) / steps
                canvas.set_pixel(int(round(py)), int(round(px)), char=char)

        # Draw the 3 edges
        draw_safe_line(x1, y1, x2, y2)
        draw_safe_line(x2, y2, x3, y3)
        draw_safe_line(x3, y3, x1, y1)


# 3. Compound Shape Class
class CompoundShape(Shape):
    def __init__(self, shapes):
        self.shapes = shapes

    def paint(self, canvas):
        for s in self.shapes:
            s.paint(canvas)
```

```python
#Testing the Canvas
my_canvas = Canvas(width=40, height=40)

# Create a house shape by compounding a Triangle (roof) and Rectangle (base)
house_base = Rectangle(length=15, width=8, x=10, y=20)
house_roof = Triangle(base=15, height=7, side1=0, side2=0, side3=0, x=10, y=20)
sun = Circle(radius=4, x=30, y=6)

my_drawing = CompoundShape([house_base, house_roof, sun])

# Paint the drawing onto the canvas
my_drawing.paint(my_canvas)

# Display the final drawing!
my_canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
#RasterDrawing Class (From Lecture 9)
class RasterDrawing:
    def __init__(self):
        self.shapes=dict()
        self.shape_names=list()
        
    def add_shape(self,shape):
        if shape.name == "":
            shape.name = self.assign_name()
        
        self.shapes[shape.name]=shape
        self.shape_names.append(shape.name)

    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)
        
    def paint(self,canvas):
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(canvas)
            
    def assign_name(self):
        name_base="shape"
        name = name_base+"_0"
        
        i=1
        while name in self.shapes:
            name = name_base+"_"+str(i)
            
        return name
```

```python
my_canvas = Canvas(width=40, height=40)
rd = RasterDrawing()

# 1. Create the shapes and give them specific names
house_base = Rectangle(length=15, width=8, x=10, y=20)
house_base.name = "base"

house_roof = Triangle(base=15, height=7, side1=0, side2=0, side3=0, x=10, y=20)
house_roof.name = "roof"

sun = Circle(radius=4, x=30, y=6)
sun.name = "sun"

# Add them to the raster drawing
rd.add_shape(house_base)
rd.add_shape(house_roof)
rd.add_shape(sun)

print("--- ORIGINAL DRAWING ---")
rd.paint(my_canvas)
my_canvas.display()

# 2.Let's make the house base wider and shift it.
print("\n--- MODIFIED DRAWING (Wider house base, same roof) ---")

rd.shapes["base"] = Rectangle(length=24, width=8, x=5, y=20)
rd.shapes["base"].name = "base"

# Update clears the canvas and repaints everything
rd.update(my_canvas)
my_canvas.display()
```

12. Implement the ability to load/save raster drawings and demonstate that your method works. One way to implement this ability:

   * Overload `__repr__` functions of all objects to return strings of the python code that would construct the object.
   
   * In the save method of raster drawing class, store the representations into the file.
   * Write a loader function that reads the file and uses `eval` to instantiate the object.

For example:

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```

```python
# 1. Overload __repr__ for all your existing Shape classes
def rect_repr(self):
    return f"Rectangle({self.get_length()}, {self.get_width()}, {self.get_x()}, {self.get_y()})"
Rectangle.__repr__ = rect_repr

def circ_repr(self):
    return f"Circle({self.get_radius()}, {self.get_x()}, {self.get_y()})"
Circle.__repr__ = circ_repr

def tri_repr(self):
# We pad the sides with 0s since they aren't used for the visual drawing
    return f"Triangle({self.get_base()}, {self.get_height()}, 0, 0, 0, {self.get_x()}, {self.get_y()})"
Triangle.__repr__ = tri_repr

def comp_repr(self):
    return f"CompoundShape({repr(self.shapes)})"
CompoundShape.__repr__ = comp_repr


# 2. Add the save method to your existing RasterDrawing class
def rd_save(self, filename):
    with open(filename, "w") as f:
        # Save both the shapes dictionary and the list of names as a tuple
        f.write(repr((self.shapes, self.shape_names)))
RasterDrawing.save = rd_save


# 3. Create the loader function
def load_drawing(filename):
    with open(filename, "r") as f:
        # eval() executes the string as python code, instantly recreating the objects
        loaded_data = eval(f.read())
        
    # Reconstruct the RasterDrawing object
    rd = RasterDrawing()
    rd.shapes = loaded_data[0]
    rd.shape_names = loaded_data[1]
    return rd
```

```python
# 4. Demonstration (Create, Save, View, Load)

print("--- 1. CREATING AND SAVING DRAWING ---")
rd_original = RasterDrawing()

house_base = Rectangle(length=15, width=8, x=10, y=20)
house_base.name = "base"
house_roof = Triangle(base=15, height=7, side1=0, side2=0, side3=0, x=10, y=20)
house_roof.name = "roof"

rd_original.add_shape(house_base)
rd_original.add_shape(house_roof)

filename = "test_drawing.txt"
rd_original.save(filename)
print(f"Drawing successfully saved to {filename}!")


print("\n--- 2. CHECKING FILE CONTENTS ---")
with open(filename, "r") as f:
    print(f.read())


print("\n--- 3. LOADING AND PAINTING FROM FILE ---")

rd_reloaded = load_drawing(filename)


new_canvas = Canvas(width=40, height=40)
rd_reloaded.update(new_canvas)


new_canvas.display()
```
